```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates


data1_path = 'Results_21Mar2022.csv'  
data2_path = 'Results_21MAR2022_nokcaladjust.csv'  

data1 = pd.read_csv(data1_path)
data2 = pd.read_csv(data2_path)


data1[['diet_group', 'sex', 'age_group']] = data1['grouping'].str.split('_', expand=True)

data1 = data1.drop(['grouping', 'mc_run_id'], axis=1)

data1['diet_group'] = data1['diet_group'].astype(str)
data1['sex'] = data1['sex'].astype(str)
data1['age_group'] = data1['age_group'].astype(str)

for col in data1.select_dtypes(include=['float64', 'int64']).columns:
    data1[col] = pd.to_numeric(data1[col], errors='coerce')
```

```python

# Create a new column to store the sum of all GHG emissions.
# GHG emissions include co2 + ch4 + n20
data1['total_ghgs'] = data1['mean_ghgs'] + data1['mean_ghgs_ch4'] + data1['mean_ghgs_n2o']

ghgs_by_diet = data1.groupby('diet_group')['total_ghgs'].mean().sort_values()

# Here I chose to use a box plot for visualization
plt.figure(figsize=(12, 8))
sns.boxplot(x='diet_group', y='total_ghgs', data=data1)
plt.title('Greenhouse Gas (co2/ch4/n2o)Emissions by Diet Habits')
plt.ylabel('Total Greenhouse Gas Emissions')
plt.xlabel('Diet Group')
plt.xticks(rotation=45)  # Rotate the diet group labels for better readability
plt.show()
```

```python

#Here in order to see how each factor behaves, I chose Parallel Coordinates Plot

numeric_cols = ['mean_land', 'mean_watscar', 'mean_eut' , 'mean_bio', 'mean_watuse','mean_acid']

plt.figure(figsize=(12, 6))
parallel_coordinates(data1, class_column='diet_group', cols=numeric_cols, colormap='viridis', linewidth=1)
plt.title('Environmental Impact by Diet Group')
plt.ylabel('Metrics')
plt.xlabel('Environmental Impact Factors')
plt.legend(title='Diet Group')
plt.show()
```

```python
#Creating a new column to store the effects of all factors other than GHGs

data1['environment_summary'] = data1['mean_land'] + data1['mean_watscar'] + data1['mean_eut']+ data1['mean_bio']+ data1['mean_watuse']+ data1['mean_acid']


ghgs_by_diet = data1.groupby('diet_group')['environment_summary'].mean().sort_values()


plt.figure(figsize=(12, 8))
sns.boxplot(x='diet_group', y='environment_summary', data=data1)
plt.title('Environmental Impact by Diet Group')
plt.ylabel('Environment Impact')
plt.xlabel('Diet Group')
plt.xticks(rotation=45)  
plt.show()
```
